```python

plt.boxplot(election.total_votes)
Q1 = election.total_votes.quantile(0.25)
Q3 = election.total_votes.quantile(0.75)
IQR = Q3 - Q1
election = election[(election.total_votes >= Q1 - 1.5*IQR) & (election.total_votes <= Q3 + 1.5*IQR)]
```

```python

plt.boxplot(election.rep_votes)
Q1 = election.rep_votes.quantile(0.25)
Q3 = election.rep_votes.quantile(0.75)
IQR = Q3 - Q1
election = election[(election.rep_votes >= Q1 - 1.5*IQR) & (election.rep_votes <= Q3 + 1.5*IQR)]
```

```python

plt.boxplot(election.dem_votes)
Q1 = election.dem_votes.quantile(0.25)
Q3 = election.dem_votes.quantile(0.75)
IQR = Q3 - Q1
election = election[(election.dem_votes >= Q1 - 1.5*IQR) & (election.dem_votes <= Q3 + 1.5*IQR)]
```

```python

plt.boxplot(election.dem_percent)
Q1 = election.dem_percent.quantile(0.25)
Q3 = election.dem_percent.quantile(0.75)
IQR = Q3 - Q1
election = election[(election.dem_percent >= Q1 - 1.5*IQR) & (election.dem_percent <= Q3 + 1.5*IQR)]
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,5))
plt1 = sns.boxplot(election['total_votes'], ax = axs[0,0])
plt2 = sns.boxplot(election['rep_votes'], ax = axs[0,1])
plt3 = sns.boxplot(election['rep_percent'], ax = axs[1,0])
plt4 = sns.boxplot(election['rep_ev'], ax = axs[1,1])

plt.tight_layout()
```
